```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("statsfinal.csv")
df
```

```python
df["total_values"] = df[["S-P1","S-P2","S-P3","S-P4"]].sum(axis=1)
df[['Date','total_values']]
```

```python
df["total_quantity"] = df[["Q-P1","Q-P2","Q-P3","Q-P4"]].sum(axis=1)
df[['Date','total_quantity']]
```

```python
df["total_revenue"] = df["total_values"] * df["total_quantity"]
```

```python
df
```

```python
max_qty = df['total_quantity'].max()
print(max_qty)
```

```python
df[df['total_quantity'] == max_qty]
```

```python
df
```

```python
df["Date"] = pd.to_datetime(df["Date"],dayfirst=True,errors="coerce")
df
```

```python
df["month"] = df["Date"].dt.month
df
```

```python
df[["Date", "month"]]
```

```python
monthly_sales = df.groupby("month")["total_values"].sum()
monthly_sales
```

```python
plt.figure(figsize=(8,5))
monthly_sales.plot(kind="bar")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.savefig("monthly_sales.png",bbox_inches="tight")
plt.show()
```

Monthly Sales Insights

```python
best_month = monthly_sales.idxmax()
best_month
```

```python
worst_month = monthly_sales.idxmin()
worst_month
```
